```python
import numpy as np
import matplotlib.pyplot as plt

print("======Dimensionamento de baterias 2======")

E = 0 #energia estocada na bateria [W*h]
U = 0 #tensão aplicada na bateria [V]
Q = 0 #capacidade da bateria [A*h]
P = 0 #potência elétrica [W]
t = 0 #tempo [h]
i = 0 #corrente [A]
#-->E = t * P
#-->E = U * Q 

#inputs sem considerar perda natural de potência do motor
i = 30
U = 22.2
P = U*i

#vamos tentar criar um gráfico da capacidade em função do tempo
Capacidade = np.array([])  #cria array vazio para as capacidades
Tempo = np.array([])       #cria array vazio para os tempos
t = 0.0001
while(t<(10/60)):
    E = t * P
    Q = (E/U)
    Tempo = np.append(Tempo, [t*60])
    Capacidade = np.append(Capacidade, [Q*1000])
    t = t + 0.005
plt.plot(Tempo, Capacidade)
plt.axis([0, 11, 0, 10000])
plt.xlabel("Tempo de voo [min]")
plt.ylabel("Capacidade da bateria [mAh]")
plt.title("Capacidade x tempo - potência = "+str(P)+"W")
plt.xticks(range(0, 11, 1))
plt.yticks(range(0, 11000, 900))
plt.grid()
plt.show()



```
